# Carparts Segmentation using Ultralytics YOLO11

This notebook serves as a starting point for training the YOLO11 model on [carparts](https://docs.ultralytics.com/datasets/segment/carparts-seg/) segmentation dataset.

## Dataset Structure

The data distribution within the Carparts Segmentation Dataset is organized as outlined below:

- **Training set**: Includes 3156 images, each accompanied by its corresponding annotations.
- **Testing set**: Comprises 276 images, with each one paired with its respective annotations.
- **Validation set**: Consists of 401 images, each having corresponding annotations.

## Applications

Carparts Segmentation finds applications in automotive quality control, auto repair, e-commerce cataloging, traffic monitoring, autonomous vehicles, insurance processing, recycling, and smart city initiatives. It streamlines processes by accurately identifying and categorizing different vehicle components, contributing to efficiency and automation in various industries.

````
import os
import glob
from collections import defaultdict

import torch
import clip
from PIL import Image

# =========================
# åŸºæœ¬é…ç½®
# =========================
device = "cuda" if torch.cuda.is_available() else "cpu"

# è½»é‡ CLIP æ¨¡å‹ï¼ˆå·²ç»æ¯”è¾ƒå°äº†ï¼‰
# ä½ ä¹Ÿå¯ä»¥æ¢æˆ "RN50" æ›´çœæ˜¾å­˜
MODEL_NAME = "RN50"

# ä½ çš„æ•°æ®é›†æ ¹ç›®å½•ï¼Œä¾‹å¦‚ï¼š
# Tesla/Tesla/cybertruck, model_3, model_s, model_x, model_y
DATASET_ROOT = "/content/Tesla2/Tesla"   # <- æ”¹è¿™é‡Œ
INDEX_PATH = "tesla_clip_index.pt"      # ç´¢å¼•æ–‡ä»¶ä¿å­˜ä½ç½®


# =========================
# 1. åŠ è½½ CLIP æ¨¡å‹
# =========================
print(f"Loading CLIP model: {MODEL_NAME} on {device}")
model, preprocess = clip.load(MODEL_NAME, device=device, jit=False)
model.eval()


# =========================
# 2. æ„å»ºç´¢å¼•ï¼ˆç¦»çº¿æ‰§è¡Œä¸€æ¬¡ï¼‰
# =========================
def build_tesla_index(dataset_root: str, index_out: str = INDEX_PATH):
    """
    éå† dataset_root ä¸‹çš„å­æ–‡ä»¶å¤¹ï¼ˆmodel_3/model_s/...ï¼‰ï¼Œ
    å¯¹æ¯å¼ å›¾æŠ½ç‰¹å¾å¹¶ä¿å­˜ä¸ºç´¢å¼•æ–‡ä»¶ã€‚
    """
    image_exts = (".jpg", ".jpeg", ".png", ".webp", ".bmp")
    all_features = []
    all_labels = []
    all_paths = []

    with torch.no_grad():
        for label in sorted(os.listdir(dataset_root)):
            class_dir = os.path.join(dataset_root, label)
            if not os.path.isdir(class_dir):
                continue

            print(f"Scanning class: {label}")
            pattern = os.path.join(class_dir, "*")
            for img_path in glob.glob(pattern):
                if not img_path.lower().endswith(image_exts):
                    continue
                try:
                    img = Image.open(img_path).convert("RGB")
                except Exception as e:
                    print(f"  [WARN] Failed to open {img_path}: {e}")
                    continue

                image_input = preprocess(img).unsqueeze(0).to(device)
                feat = model.encode_image(image_input)  # [1, D]
                feat = feat / feat.norm(dim=-1, keepdim=True)  # å½’ä¸€åŒ–

                all_features.append(feat.cpu())
                all_labels.append(label)
                all_paths.append(img_path)

    if not all_features:
        raise RuntimeError("No images found to build index, check DATASET_ROOT path.")

    features = torch.cat(all_features, dim=0)  # [N, D]
    index = {
        "features": features,       # torch.FloatTensor [N, D]
        "labels": all_labels,       # list[str], é•¿åº¦ N
        "paths": all_paths,         # list[str], é•¿åº¦ N
        "model_name": MODEL_NAME,   # è®°å½•ä¸€ä¸‹æ–¹ä¾¿ä»¥åæ£€æŸ¥
    }
    torch.save(index, index_out)
    print(f"Index saved to {index_out}, total images: {len(all_labels)}")


# =========================
# 3. åŠ è½½ç´¢å¼•
# =========================
def load_tesla_index(index_path: str = INDEX_PATH):
    index = torch.load(index_path, map_location="cpu")
    # æŠŠç‰¹å¾æ¬åˆ°å½“å‰ device ä¸Šï¼Œæ–¹ä¾¿åšç›¸ä¼¼åº¦è®¡ç®—
    index["features"] = index["features"].to(device)
    return index


# =========================
# 4. ç”¨ KNN è¿›è¡Œè½¦å‹è¯†åˆ«
# =========================
def classify_car_with_knn(
    image_path: str,
    index: dict,
    k: int = 5,
    reject_threshold: float = 0.25,
):
    """
    ä½¿ç”¨ CLIP ç‰¹å¾ + KNNï¼Œåœ¨ç´¢å¼•ä¸­æŸ¥æ‰¾æœ€ç›¸ä¼¼çš„ k å¼ å›¾ç‰‡ï¼Œ
    é€šè¿‡åŠ æƒæŠ•ç¥¨å¾—åˆ°é¢„æµ‹è½¦å‹ã€‚

    :param image_path: å¾…åˆ†ç±»å›¾ç‰‡è·¯å¾„
    :param index: å·²åŠ è½½çš„ç´¢å¼•ï¼ˆload_tesla_index çš„è¿”å›ï¼‰
    :param k: KNN çš„ k
    :param reject_threshold: å½“æœ€ä½³ç±»åˆ«çš„åŠ æƒå¾—åˆ†å¤ªä½æ—¶ï¼Œè¿”å› unknown
    :return: (pred_label, best_score, neighbors)
             neighbors ä¸º list[dict]ï¼ŒåŒ…å« top-k é‚»å±…çš„ä¿¡æ¯
    """
    with torch.no_grad():
        # 1) ç¼–ç å¾…é¢„æµ‹å›¾ç‰‡
        img = Image.open(image_path).convert("RGB")
        image_input = preprocess(img).unsqueeze(0).to(device)
        feat = model.encode_image(image_input)
        feat = feat / feat.norm(dim=-1, keepdim=True)  # [1, D]

        # 2) è®¡ç®—ä½™å¼¦ç›¸ä¼¼åº¦ï¼ˆå› ä¸ºéƒ½å½’ä¸€åŒ–äº†ï¼‰
        feats_db = index["features"]  # [N, D]
        sims = (feat @ feats_db.T).squeeze(0)  # [N]

        # 3) å– top-k é‚»å±…
        k = min(k, sims.shape[0])
        topk_vals, topk_idx = sims.topk(k, largest=True)

        labels = index["labels"]
        paths = index["paths"]

        neighbors = []
        label_scores = defaultdict(float)
        for sim, idx_i in zip(topk_vals.tolist(), topk_idx.tolist()):
            lbl = labels[idx_i]
            pth = paths[idx_i]
            label_scores[lbl] += sim
            neighbors.append({
                "label": lbl,
                "score": sim,
                "path": pth,
            })

        # 4) åŠ æƒæŠ•ç¥¨ï¼šçœ‹å“ªä¸ª label çš„æ€»ç›¸ä¼¼åº¦æœ€å¤§
        if not label_scores:
            return "unknown", 0.0, neighbors

        pred_label, best_score = max(label_scores.items(), key=lambda x: x[1])

        # 5) åŠ ä¸ªæ‹’è¯†é€»è¾‘ï¼ˆé˜ˆå€¼éœ€è¦ä½ è‡ªå·±åœ¨æ•°æ®ä¸Šè°ƒï¼‰
        if best_score < reject_threshold:
            pred_label = "unknown"

        return pred_label, float(best_score), neighbors


# =========================
# 5. ç¤ºä¾‹ï¼šå…ˆå»ºç´¢å¼•ï¼Œå†å¯¹ä¸€å¼ å›¾ç‰‡åšåˆ†ç±»
# =========================
if __name__ == "__main__":
    # ç¬¬ä¸€æ¬¡è¿è¡Œæ—¶å…ˆæ„å»ºç´¢å¼•ï¼ˆä¹‹åå¯ä»¥æ³¨é‡Šæ‰è¿™ä¸¤è¡Œï¼‰
    if not os.path.exists(INDEX_PATH):
        build_tesla_index(DATASET_ROOT, INDEX_PATH)

    # åŠ è½½ç´¢å¼•
    index = load_tesla_index(INDEX_PATH)

    # æ¢æˆä½ è¦æµ‹è¯•çš„å›¾ç‰‡è·¯å¾„
    test_img = "/content/Image (6).jpg"

    pred, score, neighbors = classify_car_with_knn(
        test_img,
        index,
        k=5,
        reject_threshold=0.25,
    )

    print(f"Query image: {test_img}")
    print(f"Predicted model: {pred} (score={score:.3f})")
    print("Top-k neighbors:")
    for n in neighbors:
        print(f"  {n['label']:10s} | {n['score']:.3f} | {n['path']}")
````
